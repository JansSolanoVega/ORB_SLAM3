```python
import cv2
import matplotlib.pyplot as plt
from PIL import Image
import os 

IMAGES_FOLDER = "dataset/KITTI-03"

IMAGE_SEQUENCE_FOLDER = os.path.join(IMAGES_FOLDER, "image_sequence")
MASK_SEQUENCE_FOLDER = os.path.join(IMAGES_FOLDER, "mask_sequence")
```

```python
def obtener_secuencias_consecutivas(lista, n):
    secuencias = []
    for i in range(len(lista) - n + 1):
        secuencia = lista[i:i+n]
        secuencias.append(secuencia)
    return secuencias

lista_original = [1, 2, 3, 4, 5, 6]
n = 3
secuencias_consecutivas = obtener_secuencias_consecutivas(lista_original, n)
print(secuencias_consecutivas)
```

```python
import re
import time
from IPython.display import display
sequence_length = 5

folder_images_path = os.path.join(IMAGES_FOLDER)
images_masks_names = os.listdir(os.path.join(folder_images_path))
image_names = []
mask_names = []
for element in images_masks_names:
    if "mask" in element:
        mask_names.append(element)
    else:
        image_names.append(element)
image_names = sorted(image_names)
mask_names = sorted(mask_names)
secuencias = obtener_secuencias_consecutivas(image_names, 5)#paths

for i, secuencia in enumerate(secuencias):
    rutaGuardarImgs = os.path.join(IMAGE_SEQUENCE_FOLDER,str(i))
    rutaGuardarMask = os.path.join(MASK_SEQUENCE_FOLDER)
    
    if not os.path.exists(rutaGuardarImgs):
        # Crear la ruta de guardar imagenes si no existe
        os.makedirs(rutaGuardarImgs)

    if not os.path.exists(rutaGuardarMask):
        # Crear la ruta de guardar mascaras si no existe
        os.makedirs(rutaGuardarMask)

    #Guardar en imagenes de secuencias de imgs
    for j, imagen in enumerate(secuencia):
        image = Image.open(os.path.join(IMAGES_FOLDER, imagen))
        #display(image)
        image.save(os.path.join(rutaGuardarImgs, str(j))+".png")
    
    #Guardar mascara para cada secuencia
    rutaMascara = os.path.join(IMAGES_FOLDER, mask_names)
    mascara = Image.open(rutaMascara)
    mascara.save(os.path.join(rutaGuardarMask, str(i))+".png")
    print(rutaMascara)
    print(rutaGuardarMask)
print(len(secuencias))
```

```python
def save_all_images_from_video(video_path, output_directory):
    # Open the video file
    video = cv2.VideoCapture(video_path)

    fps = video.get(1)
    # Initialize the frame counter
    frame_counter = 0
    # Check if the folder already exists
    if not os.path.exists(output_directory):
        # Create the folder
        os.makedirs(output_directory)
        print(f"Folder created at: {output_directory}")
    else:
        print("Folder already exists.")
        
    # Read and save all frames in the video
    while True:
        # Read the current frame
        ret, frame = video.read()

        # Break the loop if reading the frame was unsuccessful
        if not ret:
            break

        # Save the frame as an image
        image_path = f"{output_directory}/frame_{frame_counter}.jpg"
        cv2.imwrite(image_path, frame)
        print(f"Saved image: {image_path}")

        # Increment the frame counter
        frame_counter += 1

    # Release the video file and close any open windows
    video.release()
    cv2.destroyAllWindows()
```

```python
# Example usage
video_path = "videoPrueba5.mp4"
output_directory = os.path.join("E:/9noCiclo/PDM/LaneDetection/trainsetRNN/image", video_path.split(".")[0])

save_all_images_from_video(video_path, output_directory)
```
